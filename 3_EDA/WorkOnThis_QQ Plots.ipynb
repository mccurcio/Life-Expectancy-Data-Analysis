```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# For QQ plots
import statsmodels.api as sm
```

```python
path = '../data/processed/'
fileName = 'Clean_LE_Data_w_Means_2.csv'

df = pd.read_csv(path+fileName,header=0)

df.head()
```

```python
def qqplot(df):
    # Create Q-Q plot with 45-degree line
    fig = sm.qqplot(df, line='45')
    plt.show()



qqplot(df['InfD'])    
```

```python
    
import scipy.stats as stats

fig = sm.qqplot(df['InfD'], stats.t, distargs=(4,))
plt.show()
```
